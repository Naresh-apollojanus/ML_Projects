```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=200, centers=3,
                       cluster_std=0.70, random_state=0)
#y_true
#plt.scatter(X[:, 0], X[:, 1])
#y_true
```

```python
np.random.seed(1)
center_indices = np.random.choice(200, size= 3)
centers = X[center_indices]
print(centers.shape)
k = 3
```

```python
from sklearn.metrics import pairwise_distances_argmin_min
```

```python

while True:
    new_centers = []
    
    #distances = pairwise_distances(X, centers)
    closest, distances = pairwise_distances_argmin_min(X , centers) #it does two things, first calculate the distances of all point to all center thatis 600 will be the point and give choose minimum value of that data point to know in which ceenter the data point will lie.
    
    #closest = cluster assignment()
    #it will return index  1, means center 2
    plt.scatter(X[:, 0], X[:, 1], c= closest)
    plt.scatter(centers[:, 0], centers[:, 1], c= 'r')
    plt.show()
    #sum_error = np.sum(np.abs(distances))
    #print('SSE', sum_error)
    
    #print(closest)
    
    for i in range(k):
        #print('cluster', i)
        cluster_point_index = closest == i #boolean check or comparision between closest and i 
        cluster_points = X[cluster_point_index]
        #print(cluster_points)
        #print(cluster_points)
        new_center = cluster_points.mean(axis=0)
        new_centers.append(new_center)
    new_centers = np.array(new_centers)
    if np.all(new_centers == centers):
        break
    else:
        centers = new_centers
    #print(closest)
        
    
    
#print(new_centers)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_red = PCA(n_components=2)
pca_red.fit(X)
x_pca = pca_red.transform(X)


```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(X)
#clust_k.append(k)
#intra_dis.append(kmeans.inertia_)
label = kmeans.labels_
centers = kmeans.cluster_centers_
```
